```python
data.join(div_data, lsuffix='_caller', rsuffix='_other')
```

```python
df = pd.merge_ordered(data, div_data, on="Date")
```

```python
df
```

```python
payment_date = df.loc[df[df['Dividends'] > 0].index]
payment_date = payment_date[['Close', 'Dividends','Volume']]
```

```python
payment_date.reset_index()
```

```python
after_payment_date = df.loc[df[df['Dividends'] > 0].index+1]
after_payment_date = after_payment_date[['Close','Volume']]
```

This dataframe contains one row which is the dividend payment date, and below that is the day after. We now need to determine if there is any sort of affect on price/volume after a dividend is paid as a result of DRIP.
